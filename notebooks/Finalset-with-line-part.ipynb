```python
import pandas

finalset = pandas.read_csv("finalset.csv")

finalset.head()
```

```python
def match_data_get_part_line(data_list, line_part_file):
    '''opens file as dataframe finds match
    returns line and part'''

    #data_list = [aoi_x, aoi_y, aoi_width, aoi_height, aoi_token]
    
    #x	y	width	height	token
    line_dataframe = pandas.read_csv(line_part_file)
    
    line_rows = line_dataframe.index.copy()

    for row in line_rows:
        #print(line_dataframe['x'][row], data_list[0])
        
        if line_dataframe['x'][row] == data_list[0] and \
        line_dataframe['y'][row] == data_list[1] and \
        line_dataframe['width'][row] == data_list[2] and \
        line_dataframe['height'][row] == data_list[3] and \
        line_dataframe['token'][row] == data_list[4]:
            
            #line 1 part 1
            line_part = line_dataframe['name'][row]
            line = line_part.split()[1]
            part = line_part.split()[3]
            
            #print("***found***")
            return line, part
            
    print("error", data_list, line_part_file)   
    return -1, -1        


line_list = []
part_list = []


columns = finalset.columns
rows = finalset.index.copy()

#print(columns)

#go over eaver row in finalset
for row in rows:

    #find aoi_x,aoi_y, aoi_width, aoi_height from finalset set
    aoi_x = finalset['aoi_x'][row]
    aoi_y = finalset['aoi_y'][row]
    aoi_width = finalset['aoi_width'][row]
    aoi_height= finalset['aoi_height'][row]
    aoi_token = finalset['token'][row]
    
    #find code_language
    code_file = finalset['code_file'][row]

    #match found AOI values to row in code_langue + _aois_tokens.csv
    line_part_file = code_file.split('.')[0] + "_aois_tokens.csv"
    
    data = [aoi_x, aoi_y, aoi_width, aoi_height, aoi_token]
    
    line, part = match_data_get_part_line(data, line_part_file)
    
    line_list.append(line)
    part_list.append(part)
    

    
finalset["line"] = line_list
finalset["part"] = part_list 

finalset
```
