```python
import json
import matplotlib.pyplot as plt
```

```python
with open('results.json', 'r') as file:
    result = json.loads(file.read())
```

```python
fa, fp, fr, ff = result['first-order']
la, lp, lr, lf = result['linguistic']
```

```python
names = ["Nearest\nNeighbors", "Linear\nSVM", "RBF\nSVM",
         "Decision\nTree", "Random\nForest", "Neural\nNetwork", "AdaBoost",
         "Naive\nBayes"]

plt.rcParams["font.family"] = "Helvetica"

fig, ax = plt.subplots(figsize=(18,8))

ax.plot(names, fa, label='Accuracy score')
ax.plot(names, fp, linestyle='dashed', label='Precision score')
ax.plot(names, fr, linestyle='dashed', label='Recall score')
ax.plot(names, ff, linestyle='dashed', label='F1 score')

ax.scatter(names, fa)
ax.scatter(names, fp)
ax.scatter(names, fr)
ax.scatter(names, ff)
ax.legend(frameon=True)
plt.xticks(fontsize=13)

plt.ylim((0.0,1.0))


ax.set_title("Results in Subject Classification with Different Classifiers", fontsize=18);
```
