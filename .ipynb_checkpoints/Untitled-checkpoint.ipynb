```python
import pandas as pd
```

```python
datatype = {'index': str,
            'round': str, 
            'session': str,
            'line_part': str,
            'participant': str,
            'stimuli_image': str,
            'stimuli_text': str,
            'duration': 'int64',
            'x_cord': 'float64',
            'y_cord': 'float64',
            'aoi_x': 'float64',
            'aoi_y': 'float64', 
            'aoi_width': 'float64', 
            'aoi_height': 'float64',
            'token': str,
            'length': 'int64'}

df = pd.read_csv('datasets/processed_result.csv', dtype=datatype)
```

```python
df = df.drop(columns=["Unnamed: 0", "Unnamed: 0.1"])
```

```python
df['line'] = df['line_part'].apply(lambda line_part: line_part.split()[1])
```

```python
df = df.set_index('index')
```

```python
df.head()
```

```python
from tqdm import tqdm
from word import Word
```

```python
lexicon = {}

participants = {}

for row in tqdm(df.iterrows()):
    
    index = row[0]
    cols = row[1]
    
    round_id = index[0]
    session_id = index[1]
    file_id = round_id + session_id
    
    participant_id = index[2:5]
    
    line_part = cols['line_part']
    _, line_id, _, part_id = line_part.split()
    
    timestamp = cols['timestamp']
    duration = cols['duration']
    length = cols['length']
    frequency = 0
    predict = 0
    level = 0
    
    token = cols['token']
    
    key = f'{file_id}_{line_id}_{part_id}_{participant_id}'
    
    if lexicon.get(key, 0) == 0:
        lexicon[key] = Word(file_id, line_part, token)
        
    lexicon[key].add(participant=participant_id, 
                     duration=duration, 
                     start_time=timestamp, 
                     length=length, 
                     freq=frequency, 
                     predict=predict,
                     level=level)
    
    if participants.get(file_id, 0) == 0:
        participants[file_id] = []
    
    if participant_id not in participants[file_id]:
        participants[file_id].append(participant_id)
```

```python
headers = ['file',
           'line',
           'tokens',
           'subject',
           'duration',
           'SFD',
           'FFD',
           'GD',
           'TT',
           'PrF',
           'Pr1',
           'Pr2',
           'PrS',
           'length',
           'wpm']

result = pd.DataFrame(columns=headers)
```

```python
sentences = dict()

for key, word in tqdm(lexicon.items()):
    file, line, part, subject = key.split('_')
    
    sentence_key = file + '_' + line + '_' + subject

    if sentences.get(sentence_key, 0) == 0:
        sentences[sentence_key] = {}
    
    sentences[sentence_key][part] = word
```

```python
import numpy as np

for key, sentence in tqdm(sentences.items()):
    file, line, subject = key.split('_')
    
    min_word_key = np.sort(list(sentence.keys()))[0]
    max_word_key = np.sort(list(sentence.keys()))[-1]
    
    tokens = ""
    duration = sentence[max_word_key]._start_times[0] - sentence[min_word_key]._start_times[0]
    
    SFD = np.mean([word.single_fix_duration() / word.get_leng() for key, word in sentence.items()])
    FFD = np.mean([word.first_fix_duration() / word.get_leng() for key, word in sentence.items()])
    GD = np.mean([word.gaze_duration() / word.get_leng() for key, word in sentence.items()])
    TT = np.mean([word.total_time() / word.get_leng() for key, word in sentence.items()])
    PrF = np.mean([word.fix_probability() for key, word in sentence.items()])
    Pr1 = np.mean([word.prob_of_one_fix() for key, word in sentence.items()])
    Pr2 = np.mean([word.prob_of_2_or_more_fix() for key, word in sentence.items()])
    PrS = np.mean([word.prob_of_skip() for key, word in sentence.items()])
    
    length = 0
    wpm = 0
    
    values = [file, line, tokens, subject, duration, SFD, FFD, GD, TT, PrF, Pr1, Pr2, PrS, length, wpm]
    
    dic = dict(zip(headers, values))
    
    result = result.append(dic, ignore_index=True)
```
