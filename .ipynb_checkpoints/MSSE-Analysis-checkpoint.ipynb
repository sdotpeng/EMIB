```python
class Estimator:
    def __init__(self, ezreader):
        self.ezreader = ezreader
        
        with open('EZReader/datasets/1_1_corpus.txt', 'r') as file:
            self.freq_lookup = {row.strip().split()[-1] : int(row.strip().split()[0]) for row in file.readlines()}
        
            
    def fit(self):
        df = pd.read_csv('datasets/Eight_Metrics.csv')
        df = df[(df['file'] == 11) & (df['subject'] == 1)]
        
        results = [[], [], [], [], []]

        for index, row in df.iterrows():
            freq = row['freq']
            freq_class = math.floor(math.log(freq, 10)) if freq != 0 else 0
            results[freq_class].append(row['GD'])
    
        observed_GD = []
        for clss in results:
            observed_GD.append(np.array(clss).mean())
            
        return observed_GD
        
    def predict(self):
        self.ezreader.run()

        raw_prediction = self.ezreader.fetch()
        raw_prediction = raw_prediction.split('\n')
        start = raw_prediction.index(' Word-based means:') + 1
        end = raw_prediction.index(' First-fixation landing-site distributions:') - 1
        raw_prediction = raw_prediction[2 : end]
        
        results = [[], [], [], [], []]

        for line in raw_prediction:
            row = line.split()
            freq = freq_lookup[row[-1]]
            freq_class = math.floor(math.log(freq, 10)) if freq != 0 else 0
            GD = row[5]
            if GD != "Infinity":
                results[freq_class].append(int(GD))

        predicted_GD = []
        for clss in results:
            predicted_GD.append(np.array(clss).mean())
            
        return predicted_GD
        
    def score(self):

        return np.square(np.subtract(self.fit(), self.predict())).mean()
        
        
    def set_params(self, **kwargs):
        self.ezreader.set_params(**kwargs)
        
    def get_params(self):
        return self.ezreader.get_params()
```

```python
params = [{
    'a1': np.arange(100, 120, 1),
    'a2': np.arange(1.0, 3.0, 0.1),
    'a3': np.arange(5, 20, 1),
    'd': np.arange(0.01, 0.50, 0.01)
}]
```

```python
class MyGridSearch:
    def __init__(self, estimator, params):
        for a1 in params[0]['a1']:
            for a2 in params[0]['a1']:
                for a3 in params[0]['a1']:
                    for d in params[0]['a1']:
                        estimator.set_params(a1=a1, a2=a2, a3=a3, d=d)
                        print(estimator.score())
```
